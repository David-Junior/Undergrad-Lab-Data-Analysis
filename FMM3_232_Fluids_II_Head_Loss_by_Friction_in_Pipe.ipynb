```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Experimental Data
h1 = np.array([102, 100.9, 100.2, 98.8, 94.5, 93.7, 92.8, 92.5, 91.8, 92.1, 89.6, 88.5])/100 #m
h2 = np.array([71, 72.1, 72.8, 74.4, 78.1, 78.8, 79.8, 80, 80.7, 81.4, 82.9, 84])/100 #m
Vol = np.array([700, 520, 480, 440, 420, 360, 320, 320, 320, 280, 240, 220])/1000000 #m^3
t = np.array([23.5, 16.42, 16.79, 16.16, 16.61, 16.70, 17.02, 16.94, 18.42, 19.02, 16.83, 16.95]) #s

SG = 2.97
l = 2
d = 6.25/1000 #m
g = 9.81
```

```python
# Analysis
hm = h1 - h2
hf = hm * (SG-1)

A = np.pi * d**2 / 4
Q = Vol / t
v = Q / A

log_hf = np.log(hf)
log_v = np.log(v)
```

```python
# Analysed Data
DF = pd.DataFrame({'h1':h1, 'h2':h2, 'hm':hm, 'hf': hf, 'Vol':Vol, 'Q':Q, 'v': v, 'log(hf)': log_hf, 'log(v)': log_v})
DF
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(log_v, log_hf, label='Data points')

# Calculate the line of best fit (gradient and y-intercept)
gradient, y_intercept = np.polyfit(log_v, log_hf, 1)

# Create the line of best fit
line_of_best_fit = gradient * log_v + y_intercept

plt.plot(log_v, line_of_best_fit, color='red', label=f'Line of best fit (y = {gradient:.2f}x + {y_intercept:.2f})')

plt.xlabel('log(v)')
plt.ylabel('log(hf)')
plt.title('log(hf) vs log(v)')
plt.grid(True)
plt.legend()
plt.show()

print(f"Gradient of the line of best fit: {gradient:.4f}")
print(f"Y-intercept of the line of best fit: {y_intercept:.4f}")
```
