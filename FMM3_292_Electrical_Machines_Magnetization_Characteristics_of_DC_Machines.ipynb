```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Experimental data

# Field current (common to all)
If = np.array([0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5])

# 800 RPM
V_800 = np.array([0.4, 12, 22, 28, 34, 40, 42, 44])

# 1200 RPM
V_1200 = np.array([0.8, 19, 32, 44, 52, 59, 64, 68])

# 1500 RPM
V_1500 = np.array([1.4, 23, 40, 54, 66, 74, 80, 82])
```

```python
plt.figure(figsize=(8,10))
plt.plot(If, V_800, label='800 rpm')
plt.plot(If, V_1200, label='1200 rpm')
plt.plot(If, V_1500, label='1500 rpm')

plt.xlabel('Field Current I_f (A)')
plt.ylabel('Voltage E (V)')
plt.title('E vs. I_f')
plt.grid(True)
plt.xlim(0,3.5)
plt.ylim(0,90)
plt.legend()
plt.show()
```

```python
k_800 = V_800/(800*If)
k_1200 = V_1200/(1200*If)
k_1500 = V_1500/(1500*If)


plt.figure(figsize=(8,10))
plt.plot(If, k_800, label='800 rpm')
plt.plot(If, k_1200, label='1200 rpm')
plt.plot(If, k_1500, label='1500 rpm')
plt.xlabel('Field Current I_f (A)')
plt.ylabel('k')
plt.title('k against Field Current I_f')
plt.grid(True)
plt.legend()
plt.show()
```

```python
df = pd.DataFrame({
    'Field Current  I_f':If,
    'EMF (800rpm)':V_800,
    'k (800rpm)':k_800,
    'EMF (1200rpm)':V_1200,
    'k (1200rpm)':k_1200,
    'EMF (1500rpm)':V_1500,
    'k (1500rpm)':k_1500,
})
df
```

```python
# Identifying the linear region
plt.figure(figsize=(10,8))
plt.plot(If,V_1500, label = '1500 rpm')
plt.plot(If,If*38)
plt.xlim(0)
plt.ylim(0,90)
plt.grid(True)
plt.legend()
plt.xlabel('Field Current I_f (A)')
plt.ylabel('Voltage E (v)')
plt.show()

print('The range of values over which A1.2 is valid is I_f <= 1.5 A')
```
