```python
import numpy as np
```

```python
r1 = 100.31
v1 = np.array([11.93, 8.958, 5.907, 2.990])
i1_calculated = v1/r1
i1_measured = np.array([0.1188, 0.09, 0.06, 0.03])
```

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Create the scatter plot for i1_measured vs v1
plt.figure(figsize=(10, 6))
plt.scatter(v1, i1_measured,color='red')

# Perform linear regression for i1_measured vs v1
slope_measured, intercept_measured, r_value_measured, p_value_measured, std_err_measured = linregress(v1, i1_measured)

# Plot the regression line
plt.plot(v1, intercept_measured + slope_measured * v1, color='red', label=f'Measured Current: Gradient = {slope_measured:.5f}')

# Plot the calculated current (i1_calculated) vs v1
plt.plot(v1, i1_calculated, color='blue', linestyle='--', label='Calculated Current')

# Add labels, title, and legend
plt.xlabel('Voltage (v)')
plt.ylabel('Current (A)')
plt.title('Current vs Voltage for resistor 1 (100.31 Ohms)')
plt.legend()
plt.grid(True)
plt.show()
```
