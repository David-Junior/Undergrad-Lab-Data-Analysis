```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Experimental Data
r = 0.115 # radius in m
I = np.array([3.6, 4.6, 5.6, 6.6, 7.6, 8.6, 9.6, 10.6, 11.6]) # Current (A)
rpm = np.array([1798.4, 1555.4, 1287.1, 1139.6, 1032.8, 911.3, 865.4, 775.9, 716.0])
V = np.array([118, 118, 117, 117, 116, 114, 113, 112, 111]) # Voltage (V)
S1 = np.array([2, 8.5, 3.6, 3.5, 3.0, 4, 5, 6.7, 8.5 ]) # Lbs
S2 = np.array([2, 4, 11, 13, 14.9, 18.5, 20.5, 26, 31.9]) # Lbs
```

```python
# Analysis
T=abs(S1-S2)*9.81/2.20462*r # Torque
P_in=I*V # Power Input
w = rpm * 2 * np.pi / 60 # Angular Velocity in rad/s
P_out=T*w # Power Output
n= P_out / P_in * 100# Efficiency
```

```python
# Dataframe
df = pd.DataFrame({
    'Current':I,
    'Voltage':V,
    'S1':S1,
    'S2':S2,
    'S1-12':abs(S1-S2),
    'Speed':w,
    'Torque':T,
    'P_input':P_in,
    'P_output':P_out,
    'Efficiency':n
})
df
```

```python
def plot(x,y,y_label,x_label='Input Current (A)'):
  plt.figure(figsize=(10,8))
  plt.plot(x,y,'--o')
  plt.xlabel(x_label)
  plt.ylabel(y_label)
  plt.xlim(x[0])
  plt.ylim(min(y))
  plt.grid(True)
  plt.show()
```

```python
plot(I,T,'Torque (Nm)')
```

```python
plot(I, P_out, 'Power Output (W)')
```

```python
plot(I, w, 'Speed (rad/s)')
```

```python
plot(I, n, 'Efficiency (%)')
```

```python
plot(T,w,'Speed (rad/s)', 'Torque (Nm)')
```
