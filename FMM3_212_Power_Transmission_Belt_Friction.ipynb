```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- Experimental Data ---

# --- V-belt ---
v_30_t1 = np.array([22, 20, 18, 16.5, 12, 7, 1])*9.81
v_30_t2 = np.array([16, 15, 13.5, 12, 8.5, 5, 1])*9.81

v_90_t1 = np.array([22, 20.5, 18.5, 17, 11.5, 6.5, 1.5])*9.81
v_90_t2 = np.array([9, 8, 7, 6.5, 4.5, 3, 1.5])*9.81

v_150_t1 = np.array([22, 20, 18, 17, 11.5, 6, 1.5])*9.81
v_150_t2 = np.array([5, 4.5, 4, 3.5, 2.5, 1.5, 1.5])*9.81

groove = 40 # deg


# --- Flat belt ---
f_60_t1 = np.array([22, 20.5, 18.5, 17, 11.5, 6.5, 1.5])*9.81
f_60_t2 = np.array([17.5, 14.5, 13.5, 12, 8.5, 4.5, 1.5])*9.81

f_120_t1 = np.array([22, 20, 18.5, 16.5, 11.5, 6.5, 1])*9.81
f_120_t2 = np.array([11, 10.5, 9.5, 8.5, 6, 3.5, 1])*9.81

f_180_t1 = np.array([22, 19.5, 18.5, 16, 11.5, 6, 1.5])*9.81
f_180_t2 = np.array([9, 8.5, 7.5, 6.5, 4.5, 2.5, 1.5])*9.81
```

```python
plt.figure(figsize=(10,8))

r_30,c = np.polyfit(v_30_t2, v_30_t1,1)
plt.plot(v_30_t2, v_30_t2*r_30+c, color='blue')

r_90,c = np.polyfit(v_90_t2, v_90_t1,1)
plt.plot(v_90_t2, v_90_t2*r_90+c, color='orange')

r_150,c = np.polyfit(v_150_t2, v_150_t1,1)
plt.plot(v_150_t2, v_150_t2*r_150+c, color='green')

plt.scatter(v_30_t2, v_30_t1, label=f'lap angle=30, grad={r_30:.3f}',color='blue')
plt.scatter(v_90_t2, v_90_t1, label=f'lap angle=90, grad={r_90:.3f}',color='orange')
plt.scatter(v_150_t2, v_150_t1, label=f'lap angle=150, grad={r_150:.3f}',color='green')


plt.xlabel("T2 (N)")
plt.ylabel('T1 (N)')
plt.title('T1 vs T2 (v-belt)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
k_v = np.array([r_30, r_90, r_150])

theta_v = np.deg2rad([30, 90, 150])
alpha = np.deg2rad(40)

mu_v = np.log(k_v) * np.sin(alpha/2) / theta_v

print("For a v-belt, mu at different lap angles are:")
for angle, mu in zip([30, 90, 150], mu_v):
    print(f"{mu:.4f} at {angle} deg")
```

```python
plt.figure(figsize=(10,8))

r_60, c = np.polyfit(f_60_t2, f_60_t1, 1)
plt.plot(f_60_t2, f_60_t2*r_60 + c)

r_120, c = np.polyfit(f_120_t2, f_120_t1, 1)
plt.plot(f_120_t2, f_120_t2*r_120 + c)

r_180, c = np.polyfit(f_180_t2, f_180_t1, 1)
plt.plot(f_180_t2, f_180_t2*r_180 + c)


plt.scatter(f_60_t2, f_60_t1, label=f'lap angle=60, grad={r_60:.3f}')
plt.scatter(f_120_t2, f_120_t1, label=f'lap angle=120, grad={r_120:.3f}')
plt.scatter(f_180_t2, f_180_t1, label=f'lap angle=180, grad={r_180:.3f}')

plt.xlabel("T2 (N)")
plt.ylabel('T1 (N)')
plt.title('T1 vs T2 (flat-belt)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
k_f = np.array([r_60, r_120, r_180])

theta_f = np.deg2rad([60, 120, 180])

mu_f = np.log(k_f) / theta_f

print("\nFor a flat belt, mu at different lap angles are:")
for angle, mu in zip([60, 120, 180], mu_f):
    print(f"{mu:.4f} at {angle} deg")
```
